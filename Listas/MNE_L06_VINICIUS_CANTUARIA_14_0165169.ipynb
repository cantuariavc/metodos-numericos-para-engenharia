## Questão 01

O Método de Euler é o método mais direto para encontrar a solução numérica de uma Equação Diferencial Ordinária (EDO) de primeira ordem.

Considere a EDO de primeira ordem:

$$
\frac{dy}{dt} = \cos(y) + ty^2 - 5
$$

Com a condição inicial $y(0) = 1$. Dado o incremento de tempo e a quantidade de pontos, utilize o Método de Euler para encontrar o valor da função $y(t)$.

### Entrada

A entrada consiste de duas linhas. A primeira linha contém um número de ponto flutuante $\Delta t$, indicando o incremento da variável $t$. A segunda linha contém um inteiro positivo $N$, indicando o número de pontos (contando com o valor inicial) que devem ser mostrados na saída.

### Saída

A saída é composta por $N$ linhas. A $i$-ésima linha deve conter um único ponto flutuante $y_i = y(t_i) = y(i \times \Delta t)$ com 6 casas decimais de precisão.

#### Entrada de Teste
0.1  
5

#### Saída de Teste
1.000000  
0.554030  
0.142141   
-0.258464  
-0.659781

```python
# Definindo a EDO
def f(t, y):
    return np.cos(y) + t * y**2 - 5

# Leitura das entradas
delta = float(input())
N = int(input())

# Computando os resultados
y = 1
for i in range(N):
    print('%.6f' % y)
    y += delta * f(i*delta, y)
```

## Questão 02

Outro método para encontrar a solução numérica de uma EDO de primeira ordem é o Método de Euler Modificado.

Considere a EDO de primeira ordem a seguir que descreve o decaimento radioativo de determinado elemento em função do tempo:

$$
\frac{dQ(t)}{dt} = -k^2Q(t)
$$

Onde $Q(t)$ é a quantidade presente de certo elemento radioativo no instante $t$ e $k$ é uma constante que depende do elemento.

Considere $\Delta t = 1$. Dada a condição inicial $Q(0)$, a constante $k$ e a quantidade de pontos $N$, utilize o Método de Euler Modificado para encontrar o valor da função $Q(t)$.

### Entrada

A entrada consiste de três linhas. A primeira linha contém um número de ponto flutuante $Q_0 = Q(0)$. A segunda linha contém um número de ponto flutuante $k$. A terceira linha contém um inteiro positivo $N$, indicando o número de pontos (contando com o valor inicial) que devem ser mostrados na saída.

### Saída

A saída é composta por $N$ linhas. A $i$-ésima linha deve conter um único ponto flutuante $Q_i = Q(i \Delta t)$ com 6 casas decimais de precisão.

#### Entrada de Teste
1000  
1  
6

#### Saída de Teste
1000.000000  
500.000000  
250.000000  
125.000000  
62.500000  
31.250000

```python
# Definindo a EDO
def f(t, q, k):
    return -k**2 * q

# Leitura das entradas
q = float(input())
k = float(input())
N = int(input())
delta = 1

# Computando os resultados
for i in range(N):
    print('%.6f' % q)
    k1 = delta * f(i, q, k)
    k2 = delta * f(i + delta, q + k1, k)
    q += (k1 + k2) / 2
```

## Questão 03

O Método do Ponto Médio é o nome de mais um método de solução numérica de EDOs de primeira ordem. Tal método é um caso específico do conjunto de métodos de Runge-Kutta de segunda ordem.

A EDO de primeira ordem abaixo representa o modelo de Verhulst-Pearl para o crescimento de uma população em função do tempo $P(t)$.

$$
\frac{dP}{dt} = k P \begin{pmatrix}1 - \frac{P}{L}\end{pmatrix}
$$

Este modelo limita o crescimento populacional a uma constante $L$, que indica as limitações de recursos daquele local. Além disso, a constante $k$ representa a constante de crescimento da população.

Dada a condição inicial $P(0)$, o incremento de tempo $\Delta t$ e as constantes $k$ e $L$, utilize o Método de do Ponto Médio para encontrar o valor da função $P(t)$ em **10 incrementos de tempo** a partir do tempo inicial $t = 0$.

### Entrada

A entrada consiste de quatro linhas. Cada linha contém um número de ponto flutuante, na seguinte ordem: $P(0)$, $\Delta t$, $k$ e $L$.

### Saída

A saída é composta por 10 linhas. A $i$-ésima linha deve conter um único ponto flutuante $P_i = P(i \Delta t)$ com 6 casas decimais de precisão.

#### Entrada de Teste
1000  
1  
1  
100000

#### Saída de Teste
1000.000000  
2472.649750  
6015.747784  
14076.494241  
30150.742220  
54282.263056  
76496.510394  
88903.817080  
94687.620741  
97406.668561

```python
# Definindo a EDO
def f(t, p, k, L):
    return k * p * (1 - p / L)

# Leitura das entradas
p = float(input())
delta = float(input())
k = float(input())
L = float(input())
N = 10

# Computando os resultados
for i in range(N):
    print('%.6f' % p)
    k1 = delta * f(i, p, k, L)
    k2 = delta * f(i + delta/2, p + k1/2, k, L)
    p += k2
```

## Questão 04

Os Métodos de Runge-Kutta são métodos numéricos para solucionar EDOs. Tais métodos são compostos por uma sequência de etapas intermediárias e a quantidade de etapas intermediárias define a ordem do método. O conjunto de Métodos de Runge-Kutta de quarta ordem possuem quatro etapas intermediárias. Um Método de Runge-Kutta de quarta ordem bem simples e muito utilizado é o seguinte:

$$
\begin{cases}
k_1 = \Delta t f(t_n, x_n) \\
k_2 = \Delta t f(t_n + \tfrac{\Delta t}{2}, x_n + \tfrac{k_1}{2}) \\
k_3 = \Delta t f(t_n + \tfrac{\Delta t}{2}, x_n + \tfrac{k_2}{2}) \\
k_4 = \Delta t f(t_n + \Delta t, x_n + k_3) \\
x_{n+1} = x_n + \frac{1}{6}(k_1 + 2k_2 + 2k_3 + k_4)
\end{cases}
$$

A EDO de primeira ordem abaixo representa o modelo de um circuito RC ligado a uma fonte de tensão.

$$
\frac{dq}{dt} = \frac{V}{R} - \frac{q}{RC}
$$

Onde $q = q(t)$ é a quantidade de carga no tempo, $R$ é a resistência, $C$ é a capacitância e $V$ é a tensão da fonte de tensão.

Considere $q(0) = 0$ e o incremento de tempo $\Delta t = 1$. Dadas as constantes $R$, $C$ e $V$, utilize o Método de Runge-Kutta de quarta ordem mostrado acima para encontrar o valor da função $q(t)$ em **10 incrementos de tempo** a partir do tempo inicial $t = 0$.

### Entrada

A entrada consiste de três linhas. Cada linha contém um número de ponto flutuante, na seguinte ordem: $R$, $C$ e $V$.

### Saída

A saída é composta por 10 linhas. A $i$-ésima linha deve conter um único ponto flutuante $q_i = q(i \Delta t)$ com 6 casas decimais de precisão.

#### Entrada de Teste
10  
1  
5

#### Saída de Teste
0.000000  
0.475812  
0.906345  
1.295908  
1.648399  
1.967345  
2.255940  
2.517072  
2.753354  
2.967150

```python
# Definindo a EDO
def f(t, q, R, C, V):
    return V/R - q/R/C

# Leitura das entradas
R = float(input())
C = float(input())
V = float(input())
q = 0
delta = 1
N = 10

# Computando os resultados
for i in range(N):
    print('%.6f' % q)
    k1 = delta * f(i, q, R, C, V)
    k2 = delta * f(i + delta/2, q + k1/2, R, C, V)
    k3 = delta * f(i + delta/2, q + k2/2, R, C, V)
    k4 = delta * f(i + delta, q + k3, R, C, V)
    q += (k1 + 2*k2 + 2*k3 + k4) / 6
```

## Questão 05

Considere o seguinte movimento harmônico simples de um sistema massa-mola que obedeça a lei de Hooke:

$$
\frac{d^2x}{dt^2} = - \frac{kx}{m}
$$

Onde $k$ é a constante da mola e $m$ é a massa presa a ela. Dados os valores de $k$ e $m$, utilize o Método de Runge-Kutta de quarta ordem visto acima para encontrar o comportamento do corpo descrito pela função $x(t)$. Considere $\Delta t = 1$, $x(0) = 10$ e $\dot{x}(0) = 0$.

### Entrada

A entrada consiste de duas linhas. A primeira linha da entrada contém dois valores de ponto flutuante: $k$ e $m$, respectivamente. A segunda linha contém apenas um inteiro positivo $N$ indicando o número de pontos a serem calculados.

### Saída

A saída deve conter $N$ linhas. A  $i$-ésima linha deve conter um único ponto flutuante $x_i = x(iΔt)$ com 6 casas decimais de precisão.

#### Entrada de Teste
1 1  
10

#### Saída de Teste
10.000000  
5.416667  
-4.010417  
-9.695457  
-6.541733  
2.490753  
9.160549  
7.463444  
-0.963825  
-8.416886

```python
# Definindo as EDOs
# dx = y
# dy = -kx/m
def dx(t, x, y, k, m):
    return y
def dy(t, x, y, k, m):
    return -k * x / m

# Leitura das entradas
[k, m] = [float(i) for i in input().split()]
N = int(input())
x = 10
y = 0
delta = 1

# Computando os resultados
for i in range(N):
    print('%.6f' % x)

    k1 = delta * dx(i, x, y, k, m)
    l1 = delta * dy(i, x, y, k, m)
    
    k2 = delta * dx(i + delta/2, x + k1/2, y + l1/2, k, m)
    l2 = delta * dy(i + delta/2, x + k1/2, y + l1/2, k, m)
    
    k3 = delta * dx(i + delta/2, x + k2/2, y + l2/2, k, m)
    l3 = delta * dy(i + delta/2, x + k2/2, y + l2/2, k, m)
    
    k4 = delta * dx(i + delta, x + k3, y + l3, k, m)
    l4 = delta * dy(i + delta, x + k3, y + l3, k, m)

    x += (k1 + 2*k2 + 2*k3 + k4) / 6
    y += (l1 + 2*l2 + 2*l3 + l4) / 6
```

## Questão 06

O sistema massa-mola simples é constituído por um corpo de massa $m$ acoplado a uma mola com fator restaurador $k$ (constante de deformação), enquanto a outra extremidade está ligada a um ponto fixo. Quando o bloco de massa $m$ é deslocado $x_m$ metros em relação à posição inicial e solto em seguida, o sistema passa a oscilar em torno da posição de equilíbrio. A segunda lei de Newton pode ser usada para analisar tal sistema:

$$
F = m\ddot{x} = -kx
$$

Considere o sistema massa-mola acima com as seguintes condições iniciais: $x(0) = x_m$ e $\dot{x}(0) = 0$. Dados os valores de $x_m$, $m$ e $k$, utilize qualquer método numérico de solucionar EDOs para encontrar o valor de $x(10)$.

### Entrada

A entrada consiste de três linhas contendo os valores de $x_m$, $m$ e $k$, respectivamente.

### Saída

A saída deve conter uma única linha com o valor de $x(10)$ com 6 casas decimais de precisão.

#### Entrada de Teste
10  
1  
1

#### Saída de Teste
-8.444697

```python
# EDOs
def dx(t, x, v):
    return v
def dy(t, x, v):
    return -k * x / m

# Entradas e constantes
x0 = float(input())
m = float(input())
k = float(input())

N = 1000
delta = 10 / N

x,y = np.zeros(N), np.zeros(N)
x[0],y[0] = x0,0

# Computando os resultados
for i in range(N-1):
    k1 = delta * dx(i, x[i], y[i])
    l1 = delta * dy(i, x[i], y[i])

    k2 = delta * dx(i + delta/2, x[i] + k1/2, y[i] + l1/2)
    l2 = delta * dy(i + delta/2, x[i] + k1/2, y[i] + l1/2)
    
    k3 = delta * dx(i + delta/2, x[i] + k2/2, y[i] + l2/2)
    l3 = delta * dy(i + delta/2, x[i] + k2/2, y[i] + l2/2)
    
    k4 = delta * dx(i + delta, x[i] + k3, y[i] + l3)
    l4 = delta * dy(i + delta, x[i] + k3, y[i] + l3)

    x[i+1] = x[i] + (k1 + 2*k2 + 2*k3 + k4) / 6
    y[i+1] = y[i] + (l1 + 2*l2 + 2*l3 + l4) / 6

# Imprimindo os resultados
print('%.6f' % x[-1])
```

## Questão 07

$$
\begin{cases}
\frac{dx}{dt} = 10(y - x) \\
\frac{dy}{dt} = x(27 - z) - y \\
\frac{dz}{dt} = xy - \tfrac{8}{3}z
\end{cases}
$$

Tal modelo é chamado de Atrator de Lorenz. Não vamos entrar no mérito de como as equações acima foram encontradas. Considere $\{x_0, y_0, z_0\} = \{1, 2, 3\}$. Dado o incremento de tempo e a quantidade de pontos, utilize o Método de Runge-Kutta de quarta ordem para encontrar o comportamento das funções $x$, $y$ e $z$.
